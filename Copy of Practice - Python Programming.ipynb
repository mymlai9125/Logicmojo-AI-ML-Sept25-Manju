```python
n = 5
while n <= 6:
  for i in range(n):
    for j in range(n-i):
      print(" ", end="")
    for k in range(2*i+1):
      print("*", end="")
    print()
  break
while n > 0:
  for x in range(n,0,-1):
    #print(x)
    for z in range(x+1-2*(x-3)):
      print(" ", end="")
    for y in range(2*(x-1)-1):
      print("*", end="")
    print()
  break
```

```python
n=5
for i in range(n):
  if i == 0 or i == n-1:
    for j in range(n):
      print("*", end="")
  else:
    print("*", end='')
    for k in range(n-2):
      print(" ", end='')
    print("*", end='')
  print()
```

```python
s = "ashsa"
newstr = ''
i = len(s)
for i in range(len(s),0,-1):
  newstr+=s[i-1]
print(newstr)
if s == newstr:
  print("Palindrome!!")
else:
  print("Not a palindrome")
```

```python
s='aiengineer'
count=0
vowels = ['a','e','i','o','u']
for i in range(len(s)):
  for j in range(len(vowels)):
    if s[i] == vowels[j]:
      count+=1
print(count)
```

```python
s='abceio'
count=0
vowels = ['a','e','i','o','u']
for i in range(len(s)):
  if s[i] in vowels:
    count+=1
print(count)
```
